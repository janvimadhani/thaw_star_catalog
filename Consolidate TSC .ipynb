```python
def format_RA_DEC(ra_h,ra_min,dec_d,dec_min):
    
    """
    Give function RA hour, min, seconds,
                  Dec deg, min, seconds

    Get RA in format: 00 00 00.00
    Dec in format: 00 00 00.00 
    """
    

    
    ra_minute = str(ra_min)
    ra_minute = ra_minute[:2]
    ra_sec = "%.2f" % ((ra_min * 60) % 60)
    
    dec_sec = "%.2f" % ((abs(dec_min) * 60) % 60)
 
    dec_d = str(dec_d)
    
    if dec_d[0] != '-':
        dec_d = '+' + dec_d
        
    if np.isnan(ra_h) == False:
        ra_h = str(ra_h)
        if len(ra_h)  < 2 :
            ra_h = '0' + ra_h
        ra = str(ra_h) + ' ' + str(ra_minute) + ' ' + str(ra_sec) + ' '
        dec = (dec_d) + ' ' + str(dec_min) + ' ' + str(dec_sec) + ' '
        
        
    else:
        ra,dec = 0,0
 

    
    return(ra,dec)


def format_BD_number(D1,D2):

     
    if np.isnan(D1) == False:
        D1 = str(D1)
        D2 = str(D2)
        if D1[0] != '-':
            D1 = '+' + D1 
            
        bd = D1 + "˚" + D2
        
        return bd
            
    else:
        return np.nan

        
def format_milli(number):
    
    if number == '--':
        return ' '
    
    
    elif len(str(number)) > 1:
            
        number = str(number)
        
        if number[:2] == '0.':
            #print('this is option1')
            number = float(number)*1000
            return str(int(number))
        elif number[1] == '.':
            #print('this is option2')
            number = float(number)
            return str(int(number))
        elif number[0] == '.':
            #print('this is option3')
            number = float(number)*1000
            return str(int(number))
        elif (number[0] == '-') & (number[1] != '-'):
            #print('this is option 4')
            number = float(number)*1000
            return str(int(number))
        else:
            #print('this is option5')
            return(str(number)) 


def strip_trail_zeros(number):
    if number:
        if number[-2:] == '.0':
            return number[:-2]
        else:
            return number

```
