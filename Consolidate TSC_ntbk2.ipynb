### Define functions relevant for reformatting

```python
def format_RA_DEC(ra_h,ra_min,dec_d,dec_min):
    
    """
    Input parameters: RA hour, min, seconds,
                      Dec deg, min, seconds

    Returns: RA in format: 00 00 00.00
             Dec in format: 00 00 00.00 
    """
    

    
    ra_minute = str(ra_min)
    ra_minute = ra_minute[:2]
    
    if len(ra_minute) < 2:
        ra_minute = '0' + ra_minute
    
    if ra_minute[-1] == '.':
        ra_minute = '0' + ra_minute[0]
    
    ra_sec = "%.2f" % ((ra_min * 60) % 60)
    
    dec_sec = "%.2f" % ((abs(dec_min) * 60) % 60)
    dec_minute = str(dec_min)
    
    if len(dec_minute) < 2:
        dec_minute = '0' + dec_minute
    
    
    
    
        
 
    dec_d = str(dec_d)
    
    if dec_d[0] != '-':
        dec_d = '+' + dec_d
        
    if np.isnan(ra_h) == False:
        ra_h = str(ra_h)
        if len(ra_h)  < 2 :
            ra_h = '0' + ra_h
            
            
    
        ra = str(ra_h) + ' ' + str(ra_minute) + ' ' + str(ra_sec).zfill(5) + ' '
        dec = (dec_d) + ' ' + str(dec_minute) + ' ' + str(dec_sec).zfill(5) + ' '
        
        
    else:
        ra,dec = 0,0
 

    
    return(ra,dec)


def format_BD_number(D1,D2):
    """
    Input parameters: columns that have the BD number separated into
                      two components, D1 and D2
    Returns: +/- 00 0000
    """
    
    #deg = u"\u00b0" # this is the UNICODE for degree sign, might want to reinsert for latex tables 
    deg = " "

     
    if np.isnan(D1) == False:
        D1 = str(D1)
        D2 = str(D2)
        if D1[0] != '-':
            D1 = '+' + D1 
            
        bd = D1 + deg + D2
        
        return bd
            
    else:
        return np.nan
    
def format_BD_number_t4(BD_num):
    """
    Input parameters: column that has BD number starting with 'BD'
    Returns: +/- 00 0000
    """
    if BD_num:
        if BD_num[0] == 'B':
            bd = BD_num[3:]
        elif BD_num[1] == 'B':
            bd = BD_num[4:]
            
        else:
            bd = ''
    else:
        bd = ''
        
    return bd
    

        
def format_milli(number):
    """
    Input parameters: a number, the format can be any of the cases below
    Returns: milli 
    """
    
    if number == '--':
        return ' '
    
    
    elif len(str(number)) >= 1:
            
        number = str(number)
        
        if len(number) == 1:
            return(str(number))
        elif number[:2] == '0.':
            #print('this is option1')
            number = float(number)*1000
            return str(int(number))
        elif number[1] == '.':
            #print('this is option2')
            number = float(number)
            return str(int(number))
        elif number[0] == '.':
            #print('this is option3')
            number = float(number)*1000
            return str(int(number))
        elif (number[0] == '-') & (number[1] != '-'):
            #print('this is option 4')
            num = number[1:]
            #print(num)
            if len(num) <= 3:
                new_number = float(num)
            else:
                new_number = float(num)*1000
            return '-' + str(int(new_number))
        else:
            #print('this is option5')
            return(str(number)) 


def strip_trail_zeros(number):
    """
    Will remove any trailing zeros after a decimal if they exist
    Input parameters: a number that has been saved as a string
    Returns: the number (still a string_) without zeros at the end
    """
    if number:
        if number[-2:] == '.0':
            return number[:-2]
        else:
            return number

```

```python
#Write functions to turn parallax into mas and strip off the B from the beginning of BD number

def format_bd_mypc(BD_num):
    if BD_num:
        if BD_num[0] == 'B':
            bd = BD_num[2:]
        else:
            bd = ''
    else:
        bd = ''
    return bd

def format_milli_mypc(arcsec):
    if arcsec:
        if arcsec[0] == '+':
            sign = 1
            secs = float(arcsec[1:])
            milli = "%.1f" % float(sign*secs*1000)
            milli = '+' + str(milli)
        
        elif arcsec[0] == '-':
            sign = -1
            secs = float(arcsec[1:])
            milli = "%.1f" % float(sign*secs*1000)
            milli = str(milli)
        else:
            secs = float(arcsec)
            milli = "%.1f" % float(secs*1000)
            milli = str(milli)
            
    else:
        milli = ''
    return milli
```

```python
# Write a function to add together two hour angles to go from 1900 to 2000 coordinates

#first convert to minutes
def arcminutes(hourangle):
    if not ma.is_masked(hourangle):
        if len(hourangle.split()) >= 3:

            h,m,s = hourangle.split()[0],hourangle.split()[1],hourangle.split()[2] #or degree, 1hr = 60', 1 deg = 60'
            #get rid of +/- in front of dec degrees
            if h[0].isdigit():
                hour = h
            else:
                hour = h[1:]
                

            hour = float(hour)
            minute = float(m)
            sec = float(s)


            minutes = (hour*60) + (minute) + (sec/60)
            
        elif len(hourangle.split()) >= 2:
            m,s = hourangle.split()[0],hourangle.split()[1]
            
            if m[0].isdigit():
                minute = m
            else:
                minute = m[1:]
                
            minute = float(m)
            sec = float(s)
            
            minutes = (minute) + (sec/60)
        else:
            minutes = 0
    else:
        minutes = 0
    return float("%.2f" % minutes)

        
        
def add_hourangle(c1900,secval,dec=False):
    """
    parameters: 1900 coordinates (RA or Dec) [hh mm ss]
                secular variation in coordinates over 100 years (RA or Dec) [hh mm ss]
    returns: 2000 coordinates [hh mm ss]
    """
    if not ma.is_masked(c1900):
        c1900 = str(c1900)
        secval = str(secval)
        if dec == True:
            decsign = c1900[0]
        
        ### get the signs out front
        if not c1900[0].isdigit():
            sign1 = c1900[0]
            if sign1 == '-':
                sign1 = '-1'
            else:
                sign1 = '1'
            c1900 = c1900[1:]
        else:
            sign1 = '1'
        if not secval[0].isdigit():
            sign2 = secval[0]
            if sign2 == '-':
                sign2 = '-1'
            else:
                sign2 = '1'
            secval = secval[1:]
        else:
            sign2 = '1'
        
        min_1900 = arcminutes(c1900)
        min_secval = arcminutes(secval)
        
        
        min_2000 = (float(sign1)*min_1900) + (float(sign2)*min_secval)
        
        #print(min_2000)
        
        hours = "%.2f" % (float(min_2000)/60)
        
        minutes = "%.2f" % (float(min_2000) % 60)
        
        seconds = "%.2f" % ((float(min_2000) * 60) % 60)
        
        hours = str(hours)
        minutes = str(minutes)
        seconds = str(seconds)
        
        hours = hours.zfill(5)[:2]
        minutes = minutes.zfill(5)[:2]
        
        hourangle = hours + ' ' + minutes + ' '+ seconds
    
        if dec == True:
            hourangle = decsign + hours + ' ' + minutes + ' '+ seconds
        
    else:
        hourangle = ''
        
    
    
    return hourangle
    
    

# USE ASTROPY to precess to J2000 from 1900 
def convert_to_J2000(RA, Dec,start_equinox='J1900'):
    """
    Input Parameters:Give function RA 00 00 00.00,
                                   Dec 00 00 00.00
                                   Starting equinox: default = 1900
    Returns: RA in format: 00h00m00.00s
             Dec in format: 00d00m00.00s
    """
    
        
    if RA:
        ra_s = str(RA)
        dec_s = str(Dec)
        
        ra = ra_s.split()[0] + 'h' + ra_s.split()[1] + 'm' + ra_s.split()[2] + 's'
        if len(dec_s.split()) > 2:
            dec = dec_s.split()[0] + 'd' + dec_s.split()[1] + 'm' + dec_s.split()[2] + 's'
        else:
            dec = dec_s.split()[0] + 'd' + dec_s.split()[1] + 'm' + '00.00' + 's'

    
        fk0c = SkyCoord(ra, dec, frame=FK5(equinox=Time(start_equinox)))
        fk0_2000 = FK5(equinox='J2000')
        c = fk0c.transform_to(fk0_2000,type)
        string = c.to_string('hmsdms',sep=(' ',' '))
        ra2000 = string[:12]
        dec2000 = string[14:]
        
        
    else:
        ra2000,dec2000 = 0,0
 

    
    return(ra2000,dec2000)
```
